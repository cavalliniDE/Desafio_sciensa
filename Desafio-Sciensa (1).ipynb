```

file_location = "/FileStore/tables/data.json"
file_type = "json"


infer_schema = "true"
first_row_is_header = "true"
delimiter = ","


df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

```

```
display(df)
```

```
# Create a view or table

temp_table_name = "data_json"

df.createOrReplaceTempView(temp_table_name)
```

```
%sql

select * from `data_json`
limit 5
```

```
%sql
select fullVisitorId, sum(totals.pageviews) QTD from `data_json`
group by fullVisitorId
order by 2 desc
limit 10
```

```
%sql
select fullVisitorId, sum(visitNumber) QTD_sessions from `data_json`
group by fullVisitorId
order by 2 desc
limit 10
```

```
%sql
select fullVisitorId, sum(visitNumber) QTD_sessions, date from `data_json`
group by fullVisitorId,  date 
order by 2 desc
limit 10
```

```
%sql
select date, avg(totals.timeOnSite) media from `data_json`
group by date
limit 10
```

```
%sql
select fullVisitorId, sum(visitNumber) visitNumber , device.browser , date from `data_json`
group by fullVisitorId, date, device.browser
order by 2 desc
limit 10


```
